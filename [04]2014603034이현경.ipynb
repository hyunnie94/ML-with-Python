```python
def call_API(start_date, end_date):
    dt = pd.date_range(start_date, end_date)
    key = 'RS2rh6tOELAXSWgct5fX1kGUIqMoJY4d'
   # url = 'https://www.koreaexim.go.kr/site/program/financial/exchangeJSON?authkey='+key+'&searchdate='+date+'&data=AP01'
  
    headers = {'content-type': 'application/json;charset=utf-8'}
    for date in dt.strftime("%Y-%m-%d"):
        url = 'https://www.koreaexim.go.kr/site/program/financial/exchangeJSON?authkey='+key+'&searchdate='+date+'&data=AP01'

        URL = url.format(authkey=key, searchdate=date)
        response = requests.get(URL, headers=headers, verify=False)
        df = pd.DataFrame(response.json())
        result = pd.concat([df], keys=[date])
        print(result)
        result.to_csv("C:/Users/digus/Anaconda3/exchangeRate/exchange_%s.csv" % date, encoding="EUC-KR")
    return result
```

```python
input_file = r'C:\Users\digus\Anaconda3\exchangeRate' # csv파일들이 있는 디렉토리 위치
output_file = r'C:\Users\digus\Anaconda3\exchangeRate\result.csv' # 병합하고 저장하려는 파일명

allFile_list = glob.glob(os.path.join(input_file, 'exchange_*')) # glob함수로 sales_로 시작하는 파일들을 모은다
allData = []                                                    # 읽어 들인 csv파일 내용을 저장할 빈 리스트를 하나 만든다
for file in allFile_list:
    try:
        df = pd.read_csv(file, encoding='euc-kr') # for구문으로 csv파일들을 읽어 들인다
        allData.append(df)             # 빈 리스트에 읽어 들인 내용을 추가한다
    except:
        continue
dataCombine = pd.concat(allData, axis=0, ignore_index=True) # concat함수를 이용해서 리스트의 내용을 병합
# axis=0은 수직으로 병합함. axis=1은 수평. ignore_index=True는 인데스 값이 기존 순서를 무시하고 순서대로 정렬되도록 한다.
dataCombine.to_csv(output_file, index=False) # to_csv함수로 저장한다. 인데스를 빼려면 False로 설정
```

```python
from matplotlib import font_manager as fm    #시계열 데이터 그래프 
import matplotlib
%matplotlib inline
font_location = 'C:/Windows/Fonts/malgun.ttf' # For Windows
font_name = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)
exchange_rate_CNH_cut.plot()          
plt.show()

#진폭이 거의 일정한 정상 시계열
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
#송금 받을때의 데이터
plot_acf(exchange_rate_CNH_get)        #자기상관함수
plot_pacf(exchange_rate_CNH_get)       #편자기상관함수
plt.figure(figsize=(20,4))
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(exchange_rate_CNH_send)
plot_pacf(exchange_rate_CNH_send)
plt.figure(figsize=(20,4))
plt.show()
```

```python
diff_1 = exchange_rate_CNH_get.diff(periods=1).iloc[1:]
diff_1.plot()    #
plot_acf(diff_1)
plot_pacf(diff_1)
plt.show()
```

```python
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.arima_model import ARIMAResults

model = ARIMA(exchange_rate_CNH_get, order=(1, 1, 0))
model_fit = model.fit(trend = 'c', full_output = True, disp = 1)
print(model_fit.summary())
# P > |z| = 0.809 -> 거의 의미 없다
```

```python
model = ARIMA(exchange_rate_CNH_get, order=(1, 1, 0))
model_fit = model.fit(trend = 'nc', full_output = True, disp = 1) # 이 모형에 상수 포함하는 것을 제한
print(model_fit.summary())  # 아주 유의!
```

```python
model_fit.plot_predict() #ARIMA 모형이 예측한 데이터
plt.show()
```

```python
fore = model_fit.forecast(steps = 1)
print(fore) #첫번째 값을 163.2313 으로 예측
```
